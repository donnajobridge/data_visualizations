```python
import pandas as pd
import numpy as np
import seaborn as sns
import math
import matplotlib.pyplot as plt
```

```python
emo=pd.read_csv('emo.csv')
emo.head()
```

```python
accuracy=emo['response code']==emo['old/new']
```

```python
emo['accuracy']=accuracy.map({True:1, False:0})
emo['up/in']=emo['up/in'].map({'up':'Whole', 'in':'Features'})
emo['affect']=emo['affect'].map({'h':'Positive', 's':'Negative'})
```

```python
pivot=pd.pivot_table(emo, values='accuracy', index=['affect', 'up/in'], columns=['old/new'])
```

```python
means=pd.DataFrame(emo.groupby(['Subj', 'affect', 'up/in'])['accuracy'].mean())
```

```python
means.reset_index(inplace=True)
```

```python
means.melt(id_vars=['Subj', 'affect', 'up/in'])
```

```python
old=means[means['affect']!='l']
old.head()
```

```python
happy_old=old[old['affect']=='Positive']
sad_old=old[old['affect']=='Negative']
old_wide=pd.merge(happy_old, sad_old, on='Subj')
```

```python
inv = happy_old[happy_old['up/in']=='Features']
up = happy_old[happy_old['up/in']=='Whole']

up.rename(columns={'accuracy':'up_accuracy'}, inplace=True)
inv.rename(columns={'accuracy':'inv_accuracy'}, inplace=True)

up.set_index('Subj', inplace=True)
inv.set_index('Subj', inplace=True)

up=up[['up_accuracy']]
inv=inv[['inv_accuracy']]

happy_lines=pd.concat([up, inv], axis=1)
```

```python
happy_lines.T.plot()
```

```python
viosplit=sns.violinplot(x="up/in", y="accuracy", hue='affect', split=True, data=old, palette="Set2")
plt.xlabel('')
plt.ylabel('Proportion Correct')
plt.tight_layout()
plt.legend(loc='upper center')
# viosplitfig=viosplit.get_figure()
# viosplitfig.savefig('emo_viosplit.png')
```

```python
bar=sns.barplot(x="up/in", y="accuracy", hue='affect', data=old, palette="Set2")
plt.legend(loc='upper center')
plt.xlabel('')
plt.ylabel('Proportion Correct')
plt.tight_layout()

barfig=bar.get_figure()
barfig.savefig('emo_bar.png')

line_x_neg = [-.2, .8]
line_y_neg = [.672222, .672222]

line_x_pos = [.2, 1.2]
line_y_pos = [.591667, .758333]

bar.plot(line_x_pos, line_y_pos, c='orange', alpha=.5)
bar.plot(line_x_neg, line_y_neg, c='green', alpha=.5)

bar.get_xlim()
```
